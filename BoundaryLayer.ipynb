# Boundary Layers

Boundary layers arise in differential equations when the highest derivative term vanishes.  They are regions, generally close to the boundary, where the solution varies rapidly: big changes in the solution in small intervals.

Consider the differential equation $ \epsilon y'(x) + y(x) = 0 $ with bounday condition $y(0) = 1$.
The solution is simply $y(x) = exp(-x/\epsilon)$.

If we consider $\epsilon$ as a *smallness* parameter, we see that as $\epsilon \rightarrow 0$ the solution y(x) is close to the constant 0 almost anywhere, but near x close to 0 where the solution goes to 1.

```python
from math import exp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, eps):
    return exp(-x/eps)
```

```python
L = 5.
nn = 1000
dx = L/nn
xs = [n*dx for n in range(nn+1)]
```

Here you can play with your $\epsilon$ as you wish

```python
eps = 0.2
ys = [f(x, eps) for x in xs]
plt.plot(xs, ys, label=f'eps={eps}')
plt.legend()
plt.show()
```

Watch the bounday layer form as $\epsilon$ goes smaller.  Note that the **thickness** of the boundary layer is of the order of $\epsilon$ [$O(\epsilon)$].

```python
for eps in [1., 0.5, 0.2, 0.1, 0.05]:
    ys = [f(x, eps) for x in xs]
    plt.plot(xs, ys, label=f'eps={eps}')
plt.legend()
plt.title("Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.draw()
plt.savefig("bl.pdf")
```
